```python
from google.colab import files
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
uploadfile= files.upload()
df = pd.read_csv('teslastock.csv')

# Extract the 'Date' and 'Close' columns
dates = df['Date']
closing_prices = df['Close']

# Convert 'dates' to numerical values (days since the first date)
dates = pd.to_datetime(dates)  # Convert to datetime format
dates_numeric = (dates - dates.min()).dt.days  # Calculate days since the first date

# Reshape 'dates' for a single feature
dates_numeric = np.array(dates_numeric).reshape(-1, 1)

# Convert 'closing_prices' to a NumPy array
closing_prices = np.array(closing_prices)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(dates_numeric, closing_prices)

# Make predictions using the model
predictions = model.predict(dates_numeric)

# Calculate metrics
mae = mean_absolute_error(closing_prices, predictions)
mse = mean_squared_error(closing_prices, predictions)
r2 = r2_score(closing_prices, predictions)

# Visualization
plt.figure(figsize=(5, 2))
plt.scatter(dates, closing_prices, label="Actual Prices", color='blue', marker='.')
plt.plot(dates, predictions, label="Linear Regression Predictions", color='red', linestyle='-')
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Tesla Stock Price Prediction with Linear Regression")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Print metrics
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)

# Save the plot as an image (optional)
plt.savefig('tesla_stock_prediction.png')

#Show the plot (optional)
plt.show()
```

```python
from google.colab import files
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
uploadfile= files.upload()

df = pd.read_csv('teslastock.csv')

# Extract the 'Date' and 'Close' columns
dates = df['Date']
closing_prices = df['Close']

# Convert 'dates' to numerical values (days since the first date)
dates = pd.to_datetime(dates)  # Convert to datetime format
dates_numeric = (dates - dates.min()).dt.days  # Calculate days since the first date

# Normalize the closing prices to the range [0, 1]
scaler = MinMaxScaler(feature_range=(0, 1))
closing_prices_scaled = scaler.fit_transform(np.array(closing_prices).reshape(-1, 1))

# Define the number of previous days' closing prices to consider for prediction
look_back = 10

# Create sequences of data for training
X, y = [], []
for i in range(len(dates_numeric) - look_back):
    X.append(closing_prices_scaled[i:i + look_back])
    y.append(closing_prices_scaled[i + look_back])

X, y = np.array(X), np.array(y)

# Reshape the data for LSTM input (samples, time steps, features)
X = np.reshape(X, (X.shape[0], look_back, 1))

# Create the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(look_back, 1)))
model.add(LSTM(50))
model.add(Dense(1))

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Fit the model to the data
model.fit(X, y, epochs=100, batch_size=32)

# Make predictions using the model
predictions_scaled = model.predict(X)
predictions = scaler.inverse_transform(predictions_scaled)

# Calculate metrics
actual_prices = np.array(closing_prices[look_back:])
mae = mean_absolute_error(actual_prices, predictions)
mse = mean_squared_error(actual_prices, predictions)
r2 = r2_score(actual_prices, predictions)

# Visualization
plt.figure(figsize=(5, 2))
plt.plot(dates[look_back:], actual_prices, label="Actual Prices", color='blue', marker='.')
plt.plot(dates[look_back:], predictions, label="LSTM Predictions", color='red', linestyle='-')
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Tesla Stock Price Prediction with LSTM")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Print metrics
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)

# Save the plot as an image (optional)
plt.savefig('tesla_stock_prediction_lstm.png')

# Show the plot (optional)
plt.show()
```
